```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

# api-endpoint 
SEARCH_POST_URL = "https://corporation.directory/quicksearch/company"
# The search is based on the company name, city(optional) and state(optional)

dataFrame = pd.read_excel('usecase1.xlsx') #data from csv

##########################################################################
def getAddress(companyName, df):
    """
	This code fetches the company based on the company name, city and state code 
	If address not found we return bm]lank, else return the first address
	
	The companayName is filtered fromt he passed dataFrame parameter and the city and state from that record is used for web scraping
	"""
    #input_query = df.query('Company == "SOUTHCREEK MANAGEMENT COMPANY"')
    input_query = df.query('Company == "'+companyName+'"')
    #print("company name:: "+ companyName)
    company = ""
    city = ""
    state = ""
    if(len(input_query) >1):
        company = input_query.head(1).Company.item()
        city = input_query.head(1).City.item()
        state = input_query.head(1).State.item()
    else:
        company = input_query.Company.item()
        city = input_query.City.item()
        state = input_query.State.item()
    PARAMS = {'name':company, 'city':city, 'state':state, 'search': 'true'}
	# If item() is not used then the index gets prepended with the text
    # sending get request and saving the response as response object 
    resp = requests.post(url = SEARCH_POST_URL, data = PARAMS)
    soup = BeautifulSoup(resp.text,'html.parser')

    tab_list = soup.find_all('table')
    if (len(tab_list) <= 0):
        return ""

    tab_0 = tab_list[0] # First table
    addr_list = tab_0.find_all('a')
    if (len(addr_list) <= 0):
        return ""
    
    addr = addr_list[0].get_text() # First link i.e. first address
    return addr

##########################################################################
# Add a new column with the address retured from the above method
#dataFrame['NewAddress'] = dataFrame['Company'].apply(lambda c_name: getAddress(c_name, dataFrame))


```

```python
my_frame = dataFrame.head()
#my_frame['NewAddress'] = my_frame['Company'].apply(lambda c_name: getAddress(c_name, my_frame))


my_frame['NewAddress'] = my_frame['Company'].apply(lambda c_name: getAddress(c_name, my_frame))
```
